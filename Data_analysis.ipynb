```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/BA_Cleaned_review.csv')
df.sample(5)
```

```python
plt.figure(figsize = (5,4))
sns.countplot(x = 'General_sentiments',data = df)
plt.show()
```

```python
#sentiments Percentage
d = dict(df['General_sentiments'].value_counts())
d
pos = round((d['Positive']/2250)*100)
neg = round((d['Negative']/2250)*100)
nt = round((d['Neutral']/2250)*100)
print(pos,neg,nt)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize = (5,4))

fig = plt.figure()

p = sns.countplot(x = 'verified trip',data = df,palette = 'flare',hue  = 'verified trip',legend = False)
for i in p.containers:
    p.bar_label(i)
plt.title('View of verified and non-verified trips')
plt.show()

fig.set_size_inches(8,7)
fig.savefig('data/Verified and Non Verified')
```

There are <b>1898 verified Reviews</b>

And <b>352 Non-verified Reviews</b>

```python
plt.figure(figsize = (7,3))

fig = plt.figure()

p = sns.countplot(y = 'sentiments',data = df,palette = 'flare',hue  = 'sentiments',legend = False,height= 0.8)
for i in p.containers:
    p.bar_label(i)
plt.title('Types of Reviews')
plt.show()

fig.set_size_inches(10,7)
fig.savefig('data/All reviews')
```

```python
temp_df = df[['verified trip','sentiments']]
temp_df.sample()
```

```python
tab = pd.crosstab(index = temp_df['verified trip'],columns = temp_df['sentiments'])
tab
```

```python
#plt.figure(figsize = (5,4))
tab.plot(kind='bar',figsize = (5,3),color = ['#c39bd3','blue','#58d68d','#f9e79f','#85c1e9'])
plt.xlabel('Verified Trip')
plt.ylabel('Count')
plt.title('Stacked Sentiment Counts by Verified Trip')
plt.legend(title='Sentiment')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots()

# Plot on the specified axis
tab.plot(kind='bar', stacked=True, color=['#c39bd3', 'blue', '#58d68d', '#f9e79f', '#85c1e9'], ax=ax)

# Customize the plot
ax.set_xlabel('Verified Trip')
ax.set_ylabel('Count')
ax.set_title('Stacked Sentiment Counts by Verified Trip')
ax.legend(title='Sentiment')

# Set figure size and save
fig.set_size_inches(8, 7)
fig.savefig('data/Sentiment of Verified and non Verified trips.png', dpi=300)

# Show the plot
plt.show()
```
